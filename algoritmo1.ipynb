```python
# Distância Euclidiana Bidimensional
"""def DEB (px, py, qx, qy):
    d = ((px-qx)**2+(py-qy)**2)**(1/2)
    return d"""
DEB = lambda px, py, qx, qy: ((px-qx)**2+(py-qy)**2)**(1/2)


# Índice de Qualidade do Tiro
"""def IQT (qi, dm, dmax):
    indice = (1000*(qi/5)*(1-(dm/dmax)))
    return round(indice, 2)"""
IQT = lambda qi, dm, dmax: (1000*(qi/5)*(1-(dm/dmax)))


# Declarando vetores 
xv = []  # Armazena valores dos pontos das linhas(x)
yv = []  # Armazena valores dos pontos das colunas(y)

# Pegando informação do numero de acertos do atirador
QI = int(input('Informe a quantidade de tiros impactaram o alvo:'))    
print('\n')

# Se o atirador escrever mais de 5 disparos será solicitada novamente o número de disparos no alvo
if QI >= 6:
    print('O numero máximo de diparos são 5!')    
    print('\n')

# Se o atirador acertar 1 ou nenhum disparo
elif QI <= 1:
    print('Índice de qualidade do tiro: 0.00')
    print('Resultado: Não Selecionado')
    
# Se ele acertar 2 ou mais disparos
elif 6 > QI > 1:
    # Loop para coletar todas as informações dos tiros e armazenar nos vetores
    for i in range(QI):
        print(f'Tiro {i+1}:')
        x = int(input('Informe coordenada de linha(X):'))
        y = int(input('Informe coordenada de coluna(Y):'))
        print('-----------------------------------------------------\n')
        
        # Armazenando as entradas dos pontos de acerto nos vetores
        xv.append(x)
        yv.append(y)
        
        if i == 1: # Se o já obtivermos 2 pontos
            d1 = DEB(xv[i-1], yv[i-1], xv[i], yv[i])  # Utilizando a função euclidiana para medir as distâncias
            D = d1  # Acrescentando o resultado dado da função em uma outra variável
            comb = 1  # Numero de combinações até o momento
            DMmax = 45.18  # Distância média máxima para dois pontos de acerto
        elif i == 2:
            d2 = DEB(xv[i-2], yv[i-2], xv[i], yv[i])
            d3 = DEB(xv[i-1], yv[i-1], xv[i], yv[i])
            D = d1+d2+d3  # Acrescentando a soma de resultados dados pela função em uma outra variável
            comb = 3
            DMmax = 38.73
        elif i == 3:
            d4 = DEB(xv[i-3], yv[i-3], xv[i], yv[i])
            d5 = DEB(xv[i-2], yv[i-2], xv[i], yv[i])
            d6 = DEB(xv[i-1], yv[i-1], xv[i], yv[i])
            D = d1+d2+d3+d4+d5+d6
            comb = 6
            DMmax = 35.10
        elif i == 4:
            d7 = DEB(xv[i-4], yv[i-4], xv[i], yv[i])
            d8 = DEB(xv[i-3], yv[i-3], xv[i], yv[i])
            d9 = DEB(xv[i-2], yv[i-2], xv[i], yv[i])
            d10 = DEB(xv[i-1], yv[i-1], xv[i], yv[i])
            D = d1+d2+d3+d4+d5+d6+d7+d8+d9+d10
            comb = 10
            DMmax = 32.55

    DM = D/comb  # Fazendo a média onde 'D' é a soma de todas as distâncias e 'comb' é o número de combinações para o calculo da distância 
    g = IQT(QI,DM,DMmax)  # Utilizando a função para o calculo do indice de qualidade do tiro.
    faixa = round(g, 2)  # Arredondando o valor do indice para duas casas decimais.
    
    print("TIROS: 5")
    print(f"IMPACTOS: {QI}")
    
    # Condicionando as faixas de valores aceitáveis para cada resultado.
    if 700 > faixa >= 0:  
        print(f'ÍNDICE DE QUALIDADE DO TIRO: {faixa}')
        print('RESULTADO: Não Selecionado')
    
    elif 850 > faixa >= 700:  
        print(f'ÍNDICE DE QUALIDADE DO TIRO: {faixa}')
        print('RESULTADO: Nova Oportunidade')
        
    elif 1000 >= faixa >= 850:  
        print(f'ÍNDICE DE QUALIDADE DO TIRO: {faixa}')
        print('RESULTADO: Selecionado')
    
```
